```python
def convert(row):
    cn_code = pc.country_name_to_country_alpha2(row.location, cn_name_format = "default")
    conti_code = pc.country_alpha2_to_continent_code(cn_code)
    return conti_code
```

```python
vaccine_data["continent"] = vaccine_data.apply(convert, axis =1 )
```

```python
# converting continent codes to Names
conti_names = {"AS": "Asia", "SA":"South America", "AN":"Antartica", "OC":"Oceania","EU":"Europe", "NA":"North America", "AF":"Africa"}
vaccine_data["continent"] = vaccine_data["continent"].map(conti_names)
vaccine_data
```
